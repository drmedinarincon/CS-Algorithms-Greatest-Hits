```python
# This notebook applies the Huffman algorithm to find the minimun and maximun lenght of a tree/enconding
```

```python
# The file is opened and the frequencies of each node are put into the list "List"

from pathlib import Path
List= Path('huffman.txt').read_text()
List=List.split('\n')
List.pop(0)
List.pop(1000)
List=[int(x) for x in List]
print(len(List))
```

```python
# The dictionary Level_Dic counts the number of encondings needed for the node, or equivalently, the level of the
# node on the tree. The keys are the node frequencies and the values are the level in the three. All nodes start at
# level 0, but after each union in Huffman's algorithm there is an increment of +1.

Level_Dic={}
for x in List:
    Level_Dic[x]=0

# This dictionary Tree_Dic keeps track of mergers in the Huffman algorithm. The keys are the badge frequencies and 
# the values are the frequencies of the nodes in the badge. All nodes start in individual badges, but after each
# iteration of Huffman, badges are united and the resulting frequencies are added.
    
Tree_Dic={}
for x in List:
    Tree_Dic[x]=set({x})
print(Tree_Dic)
```

```python
import heapq

# A heap is used to efficiently find the nodes with lower frequencies

heapq.heapify(List)

# In each iteration the 2 nodes with lower frequencies are merged. The replacement node has for frequency the sum
# of the frequencies it is replacing. At each iteration the level of the merged nodes increases by one.

while len(Tree_Dic)>1:
    Union1=heapq.heappop(List)
    Union2=heapq.heappop(List)
    heapq.heappush(List,Union1+Union2)
    
    Tree_Dic[Union1+Union2]=Tree_Dic[Union1].union(Tree_Dic[Union2])
    Tree_Dic.pop(Union1)
    Tree_Dic.pop(Union2)
    for x in Tree_Dic[Union1+Union2]:
        Level_Dic[x]+=1
```

```python
# Maximum level / largest lenght code in Huffman's algorithm:
print( max(Level_Dic.values()))
# Lowest level / shortest lenght code in Huffman's algorithm:
print(min(Level_Dic.values()))
```
